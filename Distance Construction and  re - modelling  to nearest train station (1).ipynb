CALCULATE DISTANCES BETWEEN CONSTRUCTION AND RAILWAYS STATION 

```python
from scipy.spatial.distance import cdist
from geopy.distance import geodesic
```

```python
rail = pd.read_csv("Boston_RaiL_Network.csv")
rail.head()
```

```python
rail.to_csv("Trainstations.csv",encoding='utf-8')
```

```python
min_closest_distance=[]
min_closest_address=[]

for address, lat1, long1 in zip(df["ADDRESS"], df["lat"], df["Long"]): #A
    a = lat1, long1
    distances = []
    for name, lat2, long2 in zip(rail['FullStation'], rail["Latitude"], rail["Longitude"]): # B
        b = lat2, long2
        distances.append(geodesic(a, b).miles)
    min_distance = min(distances)
    min_index = distances.index(min_distance)
    min_closest_distance.append(min_distance)
    
    
    
    tempgeoaddress = rail['FullStation'].iloc[min_index]
    min_closest_address.append(str(tempgeoaddress))
    print(address, " is close to  ",tempgeoaddress,"[",min_index,"]", min_distance, "miles away")
```

```python
df_cons = pd.DataFrame({"ADDRESS": df["ADDRESS"],"NEIGHBOURHOOD": df["CITY"], "ZIPCODE": df["ZIPCODE"], "STATION": min_closest_address, "DISTANCE": min_closest_distance})
df_cons
```

```python
df_cons.to_csv('Constorail2.csv',encoding='utf-8')
```
